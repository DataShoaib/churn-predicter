```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
binary_cols=[col for col in cat_cols if data[col].nunique==2]

multi_cat_cols = [col for col in cat_cols if col not in binary_cols]

if multi_cat_cols:
    ct = ColumnTransformer([
        ('ohe', OneHotEncoder(drop='first'), multi_cat_cols)
    ], remainder='passthrough')

    transformed = ct.fit_transform(data)

    ohe_cols = ct.named_transformers_['ohe'].get_feature_names_out(multi_cat_cols)
    other_cols = [col for col in data.columns if col not in multi_cat_cols]
    final_cols = list(ohe_cols) + other_cols

    df_encoded = pd.DataFrame(transformed, columns=final_cols)
else:
    df_encoded = data.copy()

df_encoded.head()
```

```python
y=df_encoded["Churn"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB,BernoulliNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
from imblearn.over_sampling import SMOTE

```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
smote=SMOTE()
x_train_resampled,y_train_resampled=smote.fit_resample(x_train,y_train)
```

```python
#now equal data poins for each class
y_train.value_counts(),y_train_resampled.value_counts()
```

```python
models = {
    'Logistic Regression': LogisticRegression(),
    'Gaussian NB': GaussianNB(),
    'Bernoulli NB': BernoulliNB(),
    'SVM': SVC(),
    'KNN': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'AdaBoost': AdaBoostClassifier(),
    'Gradient Boosting': GradientBoostingClassifier()
}
```

```python
accur=[]
preci=[]
recall=[]
model_name=[]

for name,model in models.items():
      model.fit(x_train_resampled,y_train_resampled)
      pred=model.predict(x_test)
      model_name.append(name)
      accur.append(accuracy_score(y_test,pred))
      preci.append(precision_score(y_test,pred))
      recall.append(recall_score(y_test,pred))
      
    
model_df=pd.DataFrame({
    "model":model_name,
    "accuracy":accur,
    "precision": preci,
    "recall": recall
})    
model_df=model_df.sort_values(by="accuracy",ascending=False).reset_index(drop=True)
```

lets improve it
through hyperparameter tuning

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
gbc=GradientBoostingClassifier()
params = {
    "loss":["log_loss","exponential"],
    "n_estimators":[300,600,1000],
    "max_depth":[3,6,9],
    "max_features":["sqrt","log2"],
    "min_samples_split":[10,20,30],
    "learning_rate":[0.001,0.01,0.1],
    "subsample":[1.0,0.8,0.9],
    "min_samples_leaf":[2,3,5],
    "random_state":[42]    
}
```

```python
scoring = {
    'accuracy': 'accuracy',
    'recall': 'recall',
    'precision': 'precision',
    'f1': 'f1'
}
```

```python
from sklearn.model_selection import RandomizedSearchCV

rsc = RandomizedSearchCV(gbc, param_distributions=params, n_iter=100, scoring=scoring,
                         refit="recall", cv=5, n_jobs=-1, verbose=1)
rsc.fit(x_train_resampled,y_train_resampled)
```

```python
rsc.best_params_
```

```python
rsc.best_score_
```

```python
rsc.best_estimator_
```

```python
pre_after_hyp_tun=rsc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,f1_score
```

```python
print("after hyperparameter tuning accuracy :",accuracy_score(y_test,pre_after_hyp_tun))
print("after hyperparamter tuning recall : ",recall_score(y_test,pre_after_hyp_tun))
print("after hyperparamter tuning precision : ",precision_score(y_test,pre_after_hyp_tun))
print("after hyperparamter tuning f1_score : ",f1_score(y_test,pre_after_hyp_tun))
print(" ")
print(classification_report(y_test,pre_after_hyp_tun))
```

```python
print(confusion_matrix(y_test,pre_after_hyp_tun))
```

```python
best_model = rsc.best_estimator_
importances = best_model.feature_importances_
print(importances)
```

```python
feature_importance_of_each_cols = pd.DataFrame({"features_names":x_train_resampled.columns,"importance":importances})
```

```python
feature_importance_of_each_cols=feature_importance_of_each_cols.sort_values(by="importance",ascending=False).reset_index(drop=True)
```

# insight: High Churn Risk (Customer likely to churn)
Customers who are:

- On Month-to-month contracts

- Have low tenure (e.g. less than 6 months)

- Use Electronic Check as Payment Method

- And do not have Online Security

- Customers with longer tenure and long-term contracts (especially one-year and two-year) are significantly less likely to churn. 
This indicates that customer loyalty and commitment contracts reduce churn risk.
